# The Movie Database

## Introduction
The Movie Database contains data on:
- id: Identification number
- imdb_id: IMDB identification number
- popularity: Relative number of page views on The Movie Database
- budget: Budget in USD
- revenue: Revenue in USD
- original_title: Movie title
- cast: list of cast members separated by |, max five actors
- homepage: URL for the movie homepage
- director: list of directors separated by |, max five directors
- tagline: Tagline for the movie
- keywords: list of keywords associated with the movie, separated by |, max five keywords
- overview: Summary of the plot
- runtime: Movie runtime in minutes
- genres: list of genres separated by |, max five genres
- production_companies: list of production companies separated by |, max five companies
- release_date: Original release date
- vote_count: Number of votes
- vote_average: Average of votes
- release_year: Release year
- budget_adj: Budget adjusted for inflation, in 2010 US dollars
- revenue_adj: Revenue adjusted for inflation, in 2010 US dollars

## Clean
### Summary of Data Issues
#### Quality
- delete duplicate entry
- create genre_one column of the first genre in genres column
- change release_date and release_year to date datatypes

#### Tidiness
- None

```python
# Make a copy of the database
movies_df_clean = movies_df.copy()
movies_df_clean.head()
```

### Delete Duplicates

#### Define
Look at duplicate entries and delete if necessary

#### Code

```python
movies_df_clean[movies_df_clean.duplicated()]
```

```python
movies_df_clean.drop_duplicates(inplace=True)
```

#### Test

```python
movies_df_clean.duplicated().sum()
```

### Extract First Movie Genre

#### Define
Extract the string of the first movie genre, create new column genre_one

#### Code

```python
movies_df_clean['genre_one'] = movies_df_clean.genres.apply(lambda x: str(x).split('|')[0])
```

#### Test

```python
# Check genre_one column
movies_df_clean.head()
```

### release_date and release_year to datatypes

#### Define
Change release_date and release_year to date datatypes

#### Code

```python
# Change datatypes
movies_df.release_date = pd.to_datetime(movies_df.release_date)
movies_df.release_year = pd.to_datetime(movies_df.release_year)
```

```python
# Check release_date and release_year datatype
movies_df.info()
```
